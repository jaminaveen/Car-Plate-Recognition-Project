```python
## import libraries


from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import urllib.request
from urllib.request import Request, urlopen
import boto3
```

```python
# global varibales


car_images_list = []
car_plate_images_list = []
car_plate_images_text_list = []
```

```python
# SCRAPER


headers = requests.utils.default_headers()
headers.update({'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',})

for i in range(0, 9999):
    url = 'http://platesmania.com/us/gallery-' + str(i)
    req = requests.get(url,headers=headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    car_images = soup.findAll('img', attrs={'class':'lazyOwl'})
    for car in car_images:
        car_images_list.append(car['data-src'])
    
    car_plate_images = soup.findAll('img', attrs={'class':'img-responsive center-block margin-bottom-10'})
    for car in car_plate_images:
        car_plate_images_list.append(car['src'])
        car_plate_images_text_list.append(car['alt'])
    print('======================== Scraping for page',i, ' has finished ========================')
```

#### CAR NUMBER PLATE IMAGES

```python
# creating unique filename for each image in the car number plates

file_names = []
for img in car_plate_images_list:
    file_names.append(img.split('/')[-1])

file_names = ['./number_plate_images/images/'+ m + '_' + str(n) for m,n in zip(car_plate_images_text_list,file_names)]
```

```python
# downloading the images locally


for i in range(0,len(file_names)):#len(car_plate_images_list)):
    req = Request(car_plate_images_list[i], headers={"User-Agent": "Mozilla/5.0 (X11; Linux i686) AppleWebKit/537.17 (KHTML, like Gecko) Chrome/24.0.1312.27 Safari/537.17"})
    response = urlopen(req, None, 10)
    data = response.read()
    response.close()
    file_name = car_plate_images_text_list[i] 
    output_file = open(file_names[i], 'wb')
    output_file.write(data)
    output_file.close()
```

```python
# uploading car plate images to S3


aws_id = 'ENTER YOUR AWS ID HERE'
aws_secret = 'ENTER YOU SECRET KEY HERE'


client = boto3.client('s3', aws_access_key_id=aws_id, aws_secret_access_key=aws_secret)
for f in file_names[:]:
    #print(file_to_upload)
    img_name_in_s3 = f[9:]
    client.upload_file(f, 'data-pipeline-license-plate', 'License-plate/{}'.format(img_name_in_s3))
```
